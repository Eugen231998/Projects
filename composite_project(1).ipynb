–î–ª—è —Ü–µ–ª–µ–π –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–æ–¥–∞–∂ –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–π –≥–æ–¥ –¥–∞–∂–µ –≤ —Ç—Ä–∞–¥–∏—Ü–∏–æ–Ω–Ω—ã—Ö –±–∏–∑–Ω–µ—Å–∞—Ö —Ä–µ–¥–∫–æ –±–µ—Ä—É—Ç –¥–∞–Ω–Ω—ã–µ –±–æ–ª–µ–µ —á–µ–º –∑–∞ 2-3 –≥–æ–¥–∞. –ê –≤ —Ç–∞–∫–æ–π –¥–∏–Ω–∞–º–∏—á–Ω–æ –º–µ–Ω—è—é—â–µ–π—Å—è –∏–Ω–¥—É—Å—Ç—Ä–∏–∏, –∫–∞–∫ –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω—ã–µ –∏–≥—Ä—ã –∏ –≤–æ–≤—Å–µ –Ω–µ —Å—Ç–æ–∏—Ç –±—Ä–∞—Ç—å —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–π –≤—Ä–µ–º–µ–Ω–Ω–æ–π –∏–Ω—Ç–µ—Ä–≤–∞–ª - –∏–Ω–∞—á–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –∑–∞—Ö–≤–∞—Ç–∏—à—å —É–∂–µ –æ—Ç–∂–∏–≤—à–∏–µ —Ç—Ä–µ–Ω–¥—ã. –ù–æ –∏ —Å–ª–∏—à–∫–æ–º –∫–æ—Ä–æ—Ç–∫–∏–π –ø–µ—Ä–∏–æ–¥ —Ç–æ–∂–µ –±—Ä–∞—Ç—å –Ω–µ —Å—Ç–æ–∏—Ç

```python
# –î–∏–Ω–∞–º–∏–∫–∞ –ø–æ –ø–æ–ø—É–ª—è—Ä–Ω—ã–º –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º
platform_trend = actual_df.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')
top_platforms = ['PS3', 'X360', 'Wii', 'PS4', 'XOne']
platform_trend[top_platforms].plot(figsize=(12,6))
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º')
plt.xlabel('–ì–æ–¥')
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏ (–º–ª–Ω)')
plt.legend(title='–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='platform', y='total_sales', data=actual_df[actual_df['platform'].isin(top_platforms)])
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º')
plt.xlabel('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏ (–º–ª–Ω)')
plt.yscale('log')
plt.xticks(rotation=45)
plt.show()
```

```python
# –í—ã–±–æ—Ä –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º
platforms = ['PS4', 'XOne', '3DS', 'PC']

# –¶–∏–∫–ª –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º: —Ä–∞—Å—á–µ—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
for platform in platforms:
    # –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
    platform_data = actual_df[
        (actual_df['platform'] == platform) &
        actual_df['critic_score'].notna() &
        actual_df['user_score'].notna()
    ]
    
    # –†–∞—Å—á—ë—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
    corr_critic = platform_data['critic_score'].corr(platform_data['total_sales'])
    corr_user = platform_data['user_score'].corr(platform_data['total_sales'])

    print(f"\n–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞: {platform}")
    print(f"–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤: {corr_critic:.2f}")
    print(f"–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π: {corr_user:.2f}")

    # –ì—Ä–∞—Ñ–∏–∫ 1: –ü—Ä–æ–¥–∞–∂–∏ vs –û—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.regplot(x='critic_score', y='total_sales', data=platform_data, scatter_kws={'alpha':0.5})
    plt.title(f'–ü—Ä–æ–¥–∞–∂–∏ vs –û—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤ ({platform})')
    plt.xlabel('–û—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤')
    plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏ (–º–ª–Ω)')
    plt.grid(True)

    # –ì—Ä–∞—Ñ–∏–∫ 2: –ü—Ä–æ–¥–∞–∂–∏ vs –û—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
    plt.subplot(1, 2, 2)
    sns.regplot(x='user_score', y='total_sales', data=platform_data, scatter_kws={'alpha':0.5})
    plt.title(f'–ü—Ä–æ–¥–∞–∂–∏ vs –û—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ({platform})')
    plt.xlabel('–û—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
    plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏ (–º–ª–Ω)')
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```
